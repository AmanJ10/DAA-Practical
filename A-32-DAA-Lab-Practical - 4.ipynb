### Aim : <b>(Part - A)</b> 
Implement activity selection algorithm for the given scenario.             
<b>Problem Definition:</b>In the single-machine scheduling problem, we are given a set of n activities Ai. Each job i
has a starting time si, deadline di and profit pi. At any time instant, we can do only one job. Doing a job i earns a
profit pi. Generate a solution to select the largest set of mutually compatible jobs and calculate the total profit
generated by the machine. The greedy algorithm for single-machine scheduling selects the job using activity
selection algorithm.

```python
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import time
import plotly.express as px
```

```python
s = [1,3,0,5,3,5,6,8,8,2,12]
print(f"Start time list is si = {s}")
```

```python
d = [4,5,6,7,9,9,10,11,12,14,16]
print(f"Finish time list is di = {d}")
```

```python
a = ["A1","A2","A3","A4","A5","A6","A7","A8","A9","A10","A11"]
print(f"Activity name list is Ai = {a}")
```

```python
p = [10,15,14,12,20,30,32,28,30,40,45]
print(f"Profit list is pi = {p}")
```

```python
l1 = len(s)
l2 = len(d)
l3 = len(a)
l4 = len(p)
print(f"Size of all list {l1,l2,l3,l4}")
```

```python
combined_list = [(s[i], d[i], a[i], p[i]) for i in range(len(s))]
print(combined_list)

sorted_combined_list = sorted(combined_list, key=lambda x: x[1])
print(sorted_combined_list)

print(sorted_combined_list[0][2])
```

```python
k = 0
profit = p[0]
activity = []
start_time = []
end_time = []
profit_array = []

activity.append(a[0])
start_time.append(s[0])
end_time.append(d[0])
profit_array.append(p[0])

print(f"Activity : {a[0]} completed")
print(f"Profit Generated by {a[0]} : {p[0]}\n")
for m in range(1,len(s)):
    if s[m] >= d[k]:
        profit+=p[m]
        print(f"Activity : {a[m]} completed")
        print(f"Profit Generated by {a[m]} : {p[m]}\n")
        activity.append(a[m])
        start_time.append(s[m])
        end_time.append(d[m])
        profit_array.append(p[m])
        k = m

print(f"Activites Completed : {activity}")
print(f"Total Profit generarted by these activities: {profit}")
```

```python
print(start_time,end_time,profit_array,activity)
```

```python
import matplotlib.pyplot as plt
import numpy as np

tasks = activity
start_times = start_time 
durations = end_time

fig, ax = plt.subplots(figsize=(10, 5))

for i, task in enumerate(tasks):
    ax.barh(task, durations[i], left=start_times[i], height=0.5, label=task)

ax.set_xlabel('Time (seconds)')
ax.set_title('Activity Scheduling')

plt.legend()


plt.grid(axis='x')
plt.show()
```

### Aim : <b>(Part - B)</b>     
Given three stacks of the positive numbers, the task is to find the possible equal maximum sum
of the stacks with the removal of top elements allowed.

```python
stack_1 = [5,8,5,3]
stack_2 = [5,6,9,4,2,2]
stack_3 = [5,3,2,1,2]
```

```python
print(stack_1,stack_2,stack_3)
```

```python
s1 = len(stack_1)
s2 = len(stack_2)
s3 = len(stack_3)
print(f"Sizes of 3 stacks are : {s1,s2,s3}")
```

```python
def func(stack_1,stack_2,stack_3,s1,s2,s3):
    sum_1 = 0
    sum_2 = 0
    sum_3 = 0
    
    for i in range(len(stack_1)):
        sum_1 += stack_1[i]
    for i in range(len(stack_2)):
        sum_2 += stack_2[i]
    for i in range(len(stack_3)):
        sum_3 += stack_3[i]
    
    print(f"Sum of stack 1 is :{sum_1}")
    print(f"Sum of stack 2 is :{sum_2}")
    print(f"Sum of stack 3 is :{sum_3}")

    
    while(1):
        if s1 == 0 or s2 == 0 or s3 == 0:
            return 0
        
        if sum_1 == sum_2 and sum_2 == sum_3:
            return sum_1
        
        if sum_1 >= sum_2 and sum_1 >= sum_3:
            sum_1 = sum_1 - stack_1[0]
            stack_1.pop(0)
            
            
        elif sum_2 >= sum_1 and sum_2 >= sum_3:
            sum_2 = sum_2 - stack_2[0]
            stack_2.pop(0)
            
            
        elif sum_3 >= sum_1 and sum_3 >= sum_2:
            sum_3 = sum_3 - stack_3[0]
            stack_3.pop(0)
```

```python
max_sum = func(stack_1,stack_2,stack_3,s1,s2,s3)
print(f"Maximum Equal Sum of 3 Stacks is : {max_sum}")
```
